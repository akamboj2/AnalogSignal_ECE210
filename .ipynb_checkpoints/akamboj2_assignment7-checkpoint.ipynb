#### 1) Nyquist Frequency and Sampling

As we are using computers, we work in the Discrete world and not the Analog World. So we have to sample our data in order to be able to 
In order not to have aliasing, we have to sample signals at least twice the Frequency Bandwidth of the signal. Typically for human speech and musica; tracks (f=8Khz), we use 8-44.1kHz as our sampling rate in order to capture most of the bandwidth

#### 2) Fast Fourier Transform

Fast Fourier Transform (FFT) is an efficient algorithm to compute the Discrete Fourier Transform (DFT) and is widely used in signal processing applications. For more details of DFT and FFT, take ECE 310 and 311 :D
To use FFT in Python, Y_spectrum=fft.fft(signal,N)

```python
%matplotlib inline
import numpy as np
import pylab as py
import scipy as scipy
import matplotlib.pyplot as plt
import numpy.fft as fft
```

##### Sampling Frequency

Bandwidth of Sine/Cosine Wave is just the Fundamental Frequency

```python
f=300.0;        # Frequency of Signal
fs_1=4000.0;    # Oversampling
fs_2=1000.0;    
fs_3=500.0;     # Undersampling

t1=np.arange(0.0,0.02,1/fs_1)
t2=np.arange(0.0,0.02,1/fs_2)
t3=np.arange(0.0,0.02,1/fs_3)
y1=np.sin(2*np.pi*f*t1); 
y2=np.sin(2*np.pi*f*t2);
y3=np.sin(2*np.pi*f*t3);

fig, axes = plt.subplots(nrows=3, ncols=1)

axes[0].plot(t1,y1,'r.-')
axes[0].set_xlabel('t')
axes[0].set_ylabel('y')
axes[1].plot(t2,y2,'b.-')
axes[1].set_xlabel('t')
axes[1].set_ylabel('y')
axes[2].plot(t3,y3,'g.-')
axes[2].set_xlabel('t')
axes[2].set_ylabel('y')

fig.tight_layout()
plt.show()
```

#### Autocorrelation between two signals

```python
y = np.sin(2*np.pi*np.arange(0, 1, 1/1000))
t = np.arange(0, 1, 1/1000)
fig, ax = plt.subplots(nrows=1, ncols=1)
y_corr = np.correlate(y,y,"full")
y_corr = y_corr[np.size(y_corr)//2:]

ax.plot(t, y_corr, 'b--')
ax.set_xlabel('t')
ax.set_ylabel('y_corr(t)')
ax.set_title('output auto-correlation')

fig.tight_layout()
plt.show()
```

#### (a.1) generate and plot the input signal x(t)

```python
#<------- space for you to code --------->

# hint: have a variable x = x1 + x2 + ...
# hint: for-loop over n in range(3,100,2):
def f(t):
    x=0
    for n in range(3,100,2):
        x+=4/(np.pi*(n**2-4))*np.cos(n*np.pi/2*t+np.pi)
    return .5*np.cos(np.pi*t-np.pi/2)+4/(3*np.pi)*np.cos(np.pi/2*t)+x
fs=8000.0
t=np.arange(0,4*np.pi,1/fs)
y=map(f,t)
plt.plot(t,y)
plt.xlabel("time")
plt.ylabel("f(t)")
plt.title("Input signal")
print(np.shape(t),np.shape(y))
```

#### (a.2) write a function that returns the magnitude and phase of H(w) for given w

```python
# hint: replace all <text>
# hint: code this so that you can use the function in the next step as H_w_mag, H_w_phase = mag_phase_H(w) 

"""def my_function(my_arg1, my_arg2):
    my_arg1 = my_arg1 + 1
    return [my_arg1, my_arg2]

a, b = my_function(5,10)
"""
def mag_phase_H(i):
    mags=[]
    phases=[]
    for w in i:
        mag=np.pi/2*(1+w**2)
        phase= -w
        while(phase>7*np.pi/4):
            phase-=7*np.pi/4
        while(phase<-7*np.pi/4):
            phase+=7*np.pi/4
        mags.append(mag)
        phases.append(phase)
    return [mags,phases]
```

#### (a.3) plot magnitude and phase of H(w) against w

```python
w = np.linspace(-7*np.pi/4, 7*np.pi/4, 100)
H_w_mag, H_w_phase = mag_phase_H(w) 

#<------- space for you to code --------->
plt.figure()
plt.plot(w,H_w_mag)
plt.title("magnitude vs w")
plt.ylabel("magnitude")
plt.xlabel("w")

plt.figure()
plt.plot(w,H_w_phase)
plt.title("phase vs w")
plt.ylabel("phase")
plt.xlabel("w")
```

#### (a.4) plot the steady-state respnse y_ss(t)

```python
t = np.linspace(0,10,100)
y=map(f,t)
#<------- space for you to code --------->

# hint: sum up the terms (for loop not required because the range of H(w) is small)
# hint: use H_w_mag, H_w_phase = mag_phase_H(w) 

#first create an array of H(w)
H=[]
for i in range(len(H_w_mag)):
    H.append(H_w_mag[i]*np.exp(H_w_phase[i]*1j))
    

# Take the ifft of the filter H and call it h
h=np.fft.ifft(H,len(t))
#print np.shape(h)
#print np.shape(y)

# Perform linear convolution to find y_ss
y_ss=np.convolve(h,y)

#print np.shape(t)
#print np.shape(yt)
# Take the fft of this signal and plot its magnitude.
plt.plot(range(len(y_ss)),y_ss)
plt.xlabel("time")
plt.ylabel("y_ss")
plt.title("Output y_ss")
```
